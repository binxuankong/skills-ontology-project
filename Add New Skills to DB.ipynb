```python
import pandas as pd
from sqlalchemy import create_engine
from secrets import secrets
```

```python
skills = []

with open("skills/new_skills.txt") as f:
    for s in f:
        skills.append(s.strip('\n'))

skills = list(set(skills))
skills.sort()
print(skills)
```

```python
engine = create_engine(secrets['skills_db'])
df_s = pd.read_sql_query('select * from "Skill"', engine)
df_s.head()
```

```python
ex_skill = df_s.loc[df_s['Skill'].isin(skills)]['Skill'].tolist()
ex_skill
```

```python
skills_to_add = [s for s in skills if s not in ex_skill]
print(skills_to_add)
```

```python
max_id = df_s['Id'].max()
max_id
```
