```python
import pandas as pd
from difflib import SequenceMatcher

df = pd.read_csv('skills/emsi_indeed.csv').drop(columns=['Unnamed: 0'])
df.head(10)
```

```python
# Compare skill string similarity
def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
```

```python
# Split out brackets as 'Parent'
df1 = pd.DataFrame(df['Skill'].str.split('(',1).tolist(), columns=['Skill', 'Parent'])
df1['Parent'] = df1['Parent'].str[:-1]
df1.head(10)
```

```python
skills = df1['Skill'].tolist()
len(skills)
```

```python
similar_skills = {}

for i, s1 in enumerate(skills):
    for j in range(len(skills) - i - 1):
        s2 = skills[i+j+1]
        similarity = similar(s1, s2)
        if similarity > 0.7:
            similar_skills[(s1, s2)] = similarity
```
