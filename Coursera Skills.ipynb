```python
import pandas as pd
```

```python
folder = 'resource/'
file1 = folder + 'Coursera Enterprise Catalog_Master.xlsx'
file2 = folder + 'Perkeso-CADS-C4WR.xlsx'
```

### All Enterprise Courses

```python
df_course = pd.read_excel(file1, sheet_name='All Enterprise Courses_19Nov202', header=2)
df_course = df_course[['Skills Learned', 'Domain', 'Sub-Domain']]
df_course.head()
```

```python
df_course.describe(include='object')
```

```python
df_course['Domain'].unique()
```

```python
df_course['Sub-Domain'].unique()
```

```python
df_domain = df_course.groupby(['Domain', 'Sub-Domain']).count()
df_domain
```

```python
skills_dict_list = []

for i, row in df_course.iterrows():
    skills = row['Skills Learned'].split(';')
    skills = [s.strip().title() for s in skills]
    for skill in skills:
        skills_dict_list.append({'Skill': skill, 'Domain': row['Domain'], 'Sub-Domain': row['Sub-Domain']})

df_skills1 = pd.DataFrame.from_dict(skills_dict_list)
df_skills1.head()
```

```python
df_skills1.describe(include='object')
```

### Projects

```python
df_project = pd.read_excel(file1, sheet_name='Projects', header=2)
df_project = df_project[['Skills Learned', 'Domain', 'Sub-Domain']]
df_project.head()
```

```python
df_project.describe(include='object')
```

```python
skills_dict_list = []

for i, row in df_project.iterrows():
    skills = row['Skills Learned'].split(';')
    skills = [s.strip().title() for s in skills]
    for skill in skills:
        skills_dict_list.append({'Skill': skill, 'Domain': row['Domain'], 'Sub-Domain': row['Sub-Domain']})

df_skills2 = pd.DataFrame.from_dict(skills_dict_list)
df_skills2.head()
```

```python
df_skills2.describe(include='object')
```

### English

```python
df_english = pd.read_excel(file2, sheet_name='English', header=2)
df_english = df_english[['skills', 'domain', 'subdomain']]
df_english.head()
```

```python
df_english.describe(include='object')
```

```python
skills_dict_list = []

for i, row in df_english.iterrows():
    if not isinstance(row['skills'], str):
        continue
    skills = row['skills'].split(';')
    skills = [s.strip().title() for s in skills]
    for skill in skills:
        skills_dict_list.append({'Skill': skill, 'Domain': row['domain'], 'Sub-Domain': row['subdomain']})

df_skills3 = pd.DataFrame.from_dict(skills_dict_list)
df_skills3.head()
```

```python
df_skills3.describe(include='object')
```

### Consolidate

```python
df = pd.concat([df_skills1, df_skills2, df_skills3]).reset_index().drop(columns=['index'])
df.head()
```

```python
df.describe(include='object')
```

```python
# Remove 'Not Calibrated' skills
df = df.loc[df['Skill'] != 'Not Calibrated']
df.describe(include='object')
```

```python
# Drop duplicates
df = df.drop_duplicates()
df.describe(include='object')
```

```python
v = df['Skill'].value_counts()
df_dup = df[df['Skill'].isin(v.index[v.gt(5)])].sort_values(by='Skill')
df_dup.head()
```

```python
df_dup.describe(include='object')
```

```python
# Get unique skills
df_skills = df[['Skill']]
df_skills = df_skills.drop_duplicates()
df_skills.head()
```

```python
df_skills.describe(include='object')
```

```python
df.to_csv('skills/coursera_full.csv')
df_skills.to_csv('skills/coursera_skills.csv')
df_domain.to_csv('skills/coursera_domains.csv')
```
